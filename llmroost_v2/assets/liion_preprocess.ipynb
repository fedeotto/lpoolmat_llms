```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('../datasets/LiIonDatabase.xlsx')
df.head()
```

```python
# Filter out the room temperature reports
rt_df = df[(df["temperature"] >= 15) & (df["temperature"] <= 35)]
rt_df
```

```python

#taking the median of duplicated compositions
rt_df = rt_df[['composition','log_target','family','ChemicalFamily']]
# rt_df = rt_df.groupby('composition',sort=False).median().reset_index()
# rt_df.rename(columns={'composition':'formula','log_target':'target'},inplace=True)
# rt_df
```

```python
rt_df.rename(columns={'composition':'formula', 'family':'StructuralFamily','log_target':'target'},inplace=True)
```

```python
rt_df = rt_df.groupby(['formula','StructuralFamily','ChemicalFamily'],sort=False).median().reset_index()
rt_df
```

```python
rt_df = rt_df.drop_duplicates(subset='formula',keep='first')
rt_df = rt_df.reset_index(drop=True)
```

```python
rt_df.to_excel('../datasets/LiIon_roomtemp_family.xlsx',index=False)
```
